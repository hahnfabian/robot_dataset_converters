```python
import tensorflow_datasets as tfds
import tensorflow as tf
```

```python
ds = tfds.load(
    "droid",
    data_dir="gs://gresearch/robotics",
    split="train",
    shuffle_files=False
)

def filter_task(example):
    instr = example["steps"]["language_instruction"]
    return tf.strings.regex_full_match(instr, ".*drawer.*")

filtered = ds.filter(filter_task)

for ep in filtered.take(10):
    print(ep["steps"]["language_instruction"])
```
